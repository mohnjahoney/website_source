```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fib(a, b, n):
    """Using `a` and `b` as seeds, return the n'th Fibonacci number.
    
    Example:
    a = 2, b = 5, n = 6
    
    (2, 5, 2 + 5 = 7, 5 + 7 = 12, 7 + 12 = 19, 12 + 19 = 31)
    
    (2, 5, 7, 12, 19, 31)
    """
    seq = [a, b]
    for i in range(n-2):
        seq.append(seq[-1] + seq[-2])
    return seq
```

Let's test this function.

```python
a = 0
b = 1

print(fib(a, b, 10))
```

```python
a = 5
b = 6

print(fib(a, b, 10))
```

## Plotting stuff

```python
def ratio_plot(seq, ax=None):
    
    seq = np.array(seq)
    
    if ax is None:
        fig = plt.figure()
        ax = fig.gca()
    plt.plot(seq[1:] / seq[:-1], 'o-')
```

Plot a few of these curves.

```python
a = 0
b = 1

seq = fib(a, b, 10)

ratio_plot(seq)
```

Simply amazing

```python
fig = plt.figure()
ax = fig.gca()

n = 10

for a in range(5):
    for b in range(a, 5):
        seq = fib(a, b, n)
        ratio_plot(seq, ax=ax)
```
